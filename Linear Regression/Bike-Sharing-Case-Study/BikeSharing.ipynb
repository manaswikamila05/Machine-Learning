### Building model using statsmodel, for the detailed statistics

Based on the two parameters, which variable must be dropped is decided .
1. p-values represent the significance of the variables 
2. VIF represents how variables are correlated to each other

```python
# Function to calculate VIF 
# Variance Inflation Factor or VIF, gives a basic quantitative idea about how much the feature variables are correlated with each other. 

from statsmodels.stats.outliers_influence import variance_inflation_factor


def calculateVIF(df):
    df = df.drop(['const'], axis=1) # Drop the constant term 
    vif = pd.DataFrame()
    vif['Features'] = df.columns
    vif['VIF'] = [variance_inflation_factor(df.values, i) for i in range(df.shape[1])]
    vif['VIF'] = round(vif['VIF'], 2)
    vif = vif.sort_values(by = "VIF", ascending = False)
    return vif 
```

### Independence of residuals

Autocorrelation refers to the fact that observations’ errors are correlated. Durbin-Watson test is used to verify that the observations are not auto-correlated. The test will output values between 0 and 4. The closer it is to 2, the less auto-correlation there is between the various variables.

- 0 – 2: positive auto-correlation
- 2 – 4: negative auto-correlation

```python
print('The Durbin-Watson value for Final Model is',round(sm.stats.stattools.durbin_watson((y_train - y_train_pred)),4))
```

There is almost no autocorrelation.
